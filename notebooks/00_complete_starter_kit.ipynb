This guided notebook checks exact Boolean compilation, recursive composition, continual local growth, mixed geometry, and functional-memory recall. Run from a clone of the repository.

```python
%pip install -q -e ..
import numpy as np
from dendritron import BooleanDendritron, ParityTissue, boolean_cube
```

```python
cube = boolean_cube(2)
xor = BooleanDendritron.fit(cube, cube[:, 0] ^ cube[:, 1], name='xor')
parity = ParityTissue(16, xor)
rng = np.random.default_rng(42)
x = rng.integers(0, 2, size=(1000, 16))
print({'xor': xor.verify(), 'parity_16': float(np.mean(parity(x) == x.sum(axis=1) % 2)), 'nodes': parity.node_count})
```
